### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
#Converter City Miles por galão (3,6l)
```

```python
cars.rename(columns = {"City MPG" : "City_Km/Liter"}, inplace = True)

cars
```

```python
cars["City_Km/Liter"]=cars["City_Km/Liter"] * (2.35215)
cars
```

```python
cars['Decade'] = pd.cut(cars['Year'],bins=4,labels=['80s','90s','00s','10s'])
cars[['Year','Decade']]
```

```python
cars.groupby("Decade")["City_Km/Liter"].apply(lambda grps:grps.mean())
```

```python
cars.groupby(["Decade"])[['Decade',"City_Km/Liter"]].mean("City_Km/Liter")  
```

*Answer: Yes. Apparently cars are consuming more fuel over the years.*

Does the drivetrain affect fuel consumption?

```python
##Converting Highway MPG to Highway_Km/Liter
cars.rename(columns = {"Highway MPG" : "Highway_Km/Liter"}, inplace = True)

cars
```

```python
cars["Highway_Km/Liter"]=cars["Highway_Km/Liter"] * (2.35215)
cars
```

```python
#the values are different from the example

cars.groupby("Drivetrain")["Highway_Km/Liter","City_Km/Liter"].apply(lambda grps:grps.mean())
```

*Answer: Yes. Apparently 2-Wheel Drive consumes more fuel.*

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
cars['Transmission'].unique()
```

```python
cars['Transmission'] = list( map(lambda x: "Automatic" if ("Auto" in x) else "Manual",cars['Transmission']) )
cars
```

```python
consumption = cars.groupby('Transmission')['City_Km/Liter'].count()
consumption
```

```python
cars.groupby("Transmission")[["City_Km/Liter"]].mean().sort_values("City_Km/Liter",ascending=False)
```
